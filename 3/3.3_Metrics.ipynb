```python
import numpy as np
data = np.genfromtxt(r'd:\myproj\ml\3\classification.csv', delimiter=',')
```

```python
trued = data[1:,0]
predd = data[1:,1]
```

```python
ErrMat = [[0, 0], [0, 0]]

for i in range(len(trued)):
    ErrMat[int(trued[i])][int(predd[i])] += 1
 
TP = ErrMat[1][1]
FN = ErrMat[1][0]
FP = ErrMat[0][1]
TN = ErrMat[0][0]

print('>>> Error matrix <<<')
print('True positive: ', TP)
print('False positive: ', FP)
print('False negative: ', FN)
print('True negative: ', TN)
```

```python
import sklearn.metrics as skm

Accuracy_H = (TP+TN)/(TP+TN+FP+FN)
Precision_H = TP/(TP+FP)
Recall_H = TP/(TP+FN)
FM_H = (2*Precision_H*Recall_H)/(Precision_H+Recall_H)

Accuracy_SKM = skm.accuracy_score(trued, predd)
Precision_SKM = skm.precision_score(trued, predd)
Recall_SKM = skm.recall_score(trued, predd)
FM_SKM = skm.f1_score(trued, predd)

print('>>> Metrics by hand <<<')
print('Accuracy =', round(Accuracy_H, 2))
print('Precision =', round(Precision_H, 2))
print('Recall =', round(Recall_H, 2))
print('F-Measure =', round(FM_H, 2))

print('\n>>> Metrics by sklearn <<<')
print('Accuracy =', round(Accuracy_SKM, 2))
print('Precision =', round(Precision_SKM, 2))
print('Recall =', round(Recall_SKM, 2))
print('F-Measure =', round(FM_SKM, 2))
```

```python
data = np.genfromtxt(r'd:\myproj\ml\3\scores.csv', delimiter=',')
```

```python
trued = data[1:,0]
prlrd = data[1:,1]
prsvd = data[1:,2]
prmad = data[1:,3]
prdtd = data[1:,4]
```

```python
print('>>> Classificaton metrics <<<')

# LogReg
auc_lg = skm.roc_auc_score(trued, prlrd)
print('Logistic regression:', auc_lg)

# Svm
auc_svm = skm.roc_auc_score(trued, prsvd)
print('Svm: ', auc_svm)

# MethricsAlg
auc_meth = skm.roc_auc_score(trued, prmad)
print('MethrAlg: ', auc_meth)

# DecisionTree
auc_dectr = skm.roc_auc_score(trued, prdtd)
print('DecTree: ', auc_dectr)
```

```python
prc = [0, 0, 0, 0]

prc[0] = skm.precision_recall_curve(trued, prlrd)
prc[1] = skm.precision_recall_curve(trued, prsvd)
prc[2] = skm.precision_recall_curve(trued, prmad)
prc[3] = skm.precision_recall_curve(trued, prdtd)

prc_rec = [0, 0, 0, 0]

for i in range(len(prc)):
    for j in range(len(prc[i][1])):
        if (prc[i][1])[j] >= 0.7:
            if prc_rec[i] < (prc[i][0])[j]:
                prc_rec[i] = (prc[i][0])[j]

print(prc_rec)            
print(prc_rec.index(max(prc_rec)))
```
